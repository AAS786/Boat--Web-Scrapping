<a id="1"></a>  
#### <p style="padding:8px;background-color:#d5eae9;margin:0;color:#c27849;font-family:newtimeroman;font-size:150%;text-align:center;border-radius: 70px 60px;overflow:hidden;font-weight:200">Inspect Data on Website</p>

```python
soup = BeautifulSoup(htmlcontent,"html.parser")
 #print(soup)
soup.prettify    #Prettify the codes of the HTMl page
```

<a id="1"></a>  
# <p style="padding:10px;background-color:#254441;margin:0;color:#c27849;font-family:newtimeroman;font-size:150%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">TITLE OF THE PRODUCT</p> 

```python
title = soup.title
print(type(soup)) # BeautifulSoup
print(type(title)) # Tag
print(type(title.string)) # Navigablestring
```

### 1) Get the Name of the Products

```python
name=soup.title
name
```

### 2) a) Get the Price of the Products

```python
Price = soup.find_all("span",class_="price price--highlight product-card-price")
Price
```

### b) Get the Price of the Products in list

```python
Product_Price = []
for i in range(0,len(Price)):
     Product_Price.append(Price[i].get_text().strip())
Product_Price 
```

### 3) a) Get the  playback

```python
playback=soup.find_all("span",class_="metafeild-label label--highlight custom-label label ui-2")
playback

```

### 3) b) Get the  playback in list

```python
product_playback = []
for i in range(0,len(playback)):
    product_playback.append(playback[i].get_text())
product_playback
```

### 4) a) Get the  features

```python
features=soup.find_all("div",class_="features-list only-horizontal")
features
```

### 4) b) Get the  features in list

```python
product_features = []
for i in range(0,len(features)):
    product_features.append(features[i].get_text().strip())
product_features
```

### 5) a) Get the Stars of product

```python
Stars = soup.find_all("div",class_="rating__stars")
Stars
```

### 5) b) Get the Stars of product in list

```python
Review_Stars = []
for i in range(0,len(Stars)):
    Review_Stars.append(Stars[i].get_text().strip())
Review_Stars
```

### 6) a) Get the  names of product

```python
names = soup.find_all("a",class_="product-item-meta__title tile-title")
names
```

### 6) b) Get the  names of product in list

```python
product_name = []
for i in range(0,len(names)):
    product_name.append(names[i].get_text())
product_name
```

### 7) a) Get the  discount of product

```python
discount=soup.find_all("p",class_="m-0 pr-2 off")
discount
```

### 7) b) Get the  discount of product in list

```python
product_discount = []
for i in range(0,len(discount)):
    product_discount.append(discount[i].get_text())
product_discount
```

### 7) a) Get the  review of product

```python
review=soup.find_all("span",class_="rating__caption plp-reviews")
review
```

### 7) b) Get the  review of product in list

```python
product_review = []
for i in range(0,len(review)):
    product_review.append(review[i].get_text())
product_review
```

### HISTOGRAM

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.hist(x=df.Discount)
plt.xticks(rotation=90)
plt.show()
```

# BAR PLOT

```python
plt.figure(figsize=(9,9))
df["Discount"].value_counts().plot(kind="bar")
plt.title("Boat Product Discount")
plt.ylabel("Product_Name")
plt.xlabel("Discount")
plt.show()
```

# SCATTER PLOT

```python
plt.scatter(x=df["Product_Names"],y=df["Discount"])
plt.title("Boat Product Discount")
plt.ylabel("Discount")
plt.xlabel("Product_Names")
plt.xticks(rotation=90)
plt.show() 
```

```python
# Assuming df is your DataFrame
data = df['Customer_Review']

# Calculate statistics
mean_val = data.mean()
median_val = data.median()
skewness_val = skew(data)
kurtosis_val = kurtosis(data)



# Create a normal distribution curve
sns.distplot(data, fit=norm, kde=False)
plt.title('Normal Distribution Curve')
plt.xlabel(' Customer_Review')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
